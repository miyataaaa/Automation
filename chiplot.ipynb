```python
import os
import sys
import numpy as np
import pandas as pd
import xlwings as xw

dirpath = r"C:\Users\miyar\OneDrive\デスクトップ\四年\VBA\automation\m1_n1.95"
fname = "compiled_m=1.0_n=1.95" + ".xlsx"
fpath = os.path.join(dirpath, fname)
book = xw.Book(fpath)
sheets = book.sheets
sheet2 = sheets[0]
df_sheets = pd.read_excel(fpath, sheet2.name, index_col=None)
book.close()
df_sheets
```

```python
df_sheets.shape
```

```python
# df_sheets = df_sheets.fillna(0)
from sklearn import linear_model
clf = linear_model.LinearRegression(fit_intercept=False)

coefs = []
r2s = []
for i in range(0, df_sheet.shape[1], 2):
#     print(i)
    chi = df_sheet.iloc[:, i].to_numpy()
    z = df_sheet.iloc[:, i+1].to_numpy()
    nan_row = np.where(np.isnan(chi))[0]
    if len(nan_row) != 0:
        chi = chi[:nan_row[0]].reshape(-1, 1)
        z = z[:nan_row[0]]
        clf.fit(chi, z)
        coefs.append(clf.coef_[0])
        r2s.append(clf.score(chi, z))
    else:
#         chi.reshape(-1, 1)
        clf.fit(chi.reshape(-1, 1), z)
        coefs.append(clf.coef_[0])
        r2s.append(clf.score(chi.reshape(-1, 1), z))
print("coefs: \n{}\nr2s: \n{}".format(coefs, r2s))
```

```python
# 回帰係数
print(clf.coef_[0])
 
# 切片 (誤差)
print(clf.intercept_)
 
# 決定係数
print(clf.score(chi, z))
```

```python
# matplotlib パッケージを読み込み
import matplotlib.pyplot as plt
 
# 散布図
plt.scatter(chi, z)
 
# 回帰直線
plt.plot(chi, clf.predict(chi))
```
